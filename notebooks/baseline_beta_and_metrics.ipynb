## Prettify a bit

```python
drop_cols = [
    "key",
    "audio_path",
    "album_id",
    "artist_id",
    "track_id",
    
]
features = df.drop(columns=drop_cols)
features.head()
```

```python
features = features.dropna(subset="track_name")
features.head()
```

```python
features["artist_name"] = features["artist_name"].apply(set)
features.head()
```

```python
bad_date = ~features["album_release_date"].str.fullmatch("\d{4}-\d{2}-\d{2}")
features.loc[bad_date, "album_release_date"] = features.loc[bad_date, "album_release_date"].str[:10]
```

```python
bad_date = ~features["album_release_date"].str.fullmatch("\d{4}-\d{2}-\d{2}")
# костыль для избаваления от дат типа: 1970-01-01-01 (такие есть)
features.loc[bad_date, "album_release_date"] = features.loc[bad_date, "album_release_date"].str[:10]
features["album_year"] = pd.to_datetime(features["album_release_date"]).dt.year
model_columns = features.columns.difference(["album_name", "artist_name", "track_name", "album_release_date", "genres"])
model_features = features[model_columns]
model_features.head()
```

## Simplest baseline: Normalization + Genre onehot + NearestNeighbors

```python
from sklearn.compose import ColumnTransformer  # noqa
from sklearn.impute import SimpleImputer  # noqa
from sklearn.preprocessing import OneHotEncoder, StandardScaler  # noqa
from sklearn.pipeline import make_pipeline  # noqa
from sklearn.neighbors import NearestNeighbors  # noqa
```

```python
model_features[["explicit", "is_local"]] = model_features[["explicit", "is_local"]].astype("int")
```

```python
from sklearn import set_config  # noqa


set_config(display="diagram")

cat_features = model_features.select_dtypes("object").columns
num_features = model_features.select_dtypes("number").columns

cat_encoder = OneHotEncoder(handle_unknown="ignore")
imputer = SimpleImputer()
scaler = StandardScaler()
neighbors_model = NearestNeighbors(n_neighbors=4)  ## itself + 3

preprocessor = ColumnTransformer([
    ("num_transformer", make_pipeline(imputer, scaler), num_features),
    ("cat_transformer", cat_encoder, cat_features),
])

preprocessor
```

```python
preprocessor.fit(model_features)
prepared_features = pd.DataFrame(
    preprocessor.transform(model_features).toarray(), columns=preprocessor.get_feature_names_out()
)
prepared_features.head()
```

```python
neighbors_model.fit(prepared_features)
neighbors = neighbors_model.kneighbors(prepared_features, return_distance=False)
```

```python
neighbors_indexes = neighbors[:, 1:]
neighbors_indexes[:5]
```

```python
base_indexes = [[i] * 3 for i in range(features.shape[0])]
base_indexes[:5]
```

```python
import numpy as np  # noqa


base_tracks = features.take(np.ravel(base_indexes))
neighbor_tracks = features.take(np.ravel(neighbors_indexes))
neighbor_tracks = neighbor_tracks.rename(columns={
    col: f"{col}_neighbor" for col in neighbor_tracks.columns
})
df_concat = pd.concat([base_tracks.reset_index(drop=True), neighbor_tracks.reset_index(drop=True)], axis=1)
df_concat.head()
```

```python
take_cols = [
    "genre",
    "album_name",
    "album_release_date",
    "artist_name",
    "track_name",
]
take_cols_neigh = [f"{col}_neighbor" for col in take_cols]
df_result = df_concat[take_cols + take_cols_neigh]
```

```python
check_genres = ["hip-hop", "pop", "rock"]
df_result[df_result.genre.isin(check_genres)].iloc[150:].head(10)
```
