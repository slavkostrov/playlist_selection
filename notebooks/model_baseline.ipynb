## Get data from S3

```python
import boto3  # noqa

HOST = "storage.yandexcloud.net"
SCHEMA = "https"
AWS_ACCESS_KEY_ID = "..."
AWS_SECRET_ACCESS_KEY = "..."
BUCKET_NAME = "hse-project-playlist-selection"

s3_client = boto3.client(
    "s3",
    aws_access_key_id=AWS_ACCESS_KEY_ID,
    aws_secret_access_key=AWS_SECRET_ACCESS_KEY,
    endpoint_url=f"{SCHEMA}://{HOST}",
)
```

```python
from src.tracks import S3Dataset  # noqa

s3_dataset = S3Dataset(s3_client=s3_client, bucket_name=BUCKET_NAME, prefix="tracks")
s3_dataset.scan()
```

```python
import pandas as pd  # noqa: E402

pd.set_option("display.max_columns", 200)
df = s3_dataset.to_pandas()
df.shape
```

```python
df.head()
```

## Pretify dataset

```python
drop_cols = [
    "key",
    "audio_path",
    "album_id",
    "artist_id",
    "track_id",
]
features = df.drop(columns=drop_cols)
features.head()
```

```python
features = features.dropna(subset="track_name")
features.head()
```

```python
features["genres"] = features["genres"].apply(eval)
features["artist_name"] = features["artist_name"].apply(eval)
```

```python
features["artist_name"] = features["artist_name"].apply(set)
features.head()
```

```python
bad_date = ~features["album_release_date"].str.fullmatch("\d{4}-\d{2}-\d{2}")
#      : 1970-01-01-01 ( )
features.loc[bad_date, "album_release_date"] = features.loc[bad_date, "album_release_date"].str[:10]
features["album_year"] = pd.to_datetime(features["album_release_date"]).dt.year
model_columns = features.columns.difference(["album_name", "artist_name", "track_name", "album_release_date",  "genres"])
model_features = features[model_columns]
model_features.head()
```

## Feature preprocessing

```python
import numpy as np  # noqa
from sklearn.compose import ColumnTransformer  # noqa
from sklearn.impute import SimpleImputer  # noqa
from sklearn.preprocessing import OneHotEncoder, StandardScaler  # noqa
from sklearn.pipeline import make_pipeline  # noqa
from sklearn.neighbors import NearestNeighbors  # noqa
from sklearn.decomposition import PCA  # noqa
```

```python
model_features[["explicit", "is_local"]] = model_features[["explicit", "is_local"]].astype("int")
```

```python
from sklearn import set_config  # noqa


set_config(display="diagram")

cat_features = model_features.select_dtypes("object").columns
num_features = model_features.select_dtypes("number").columns

cat_encoder = OneHotEncoder(handle_unknown="ignore")
imputer = SimpleImputer()
scaler = StandardScaler()
neighbors_model = NearestNeighbors(n_neighbors=4, metric="manhattan")  ## itself + 3


preprocessor = ColumnTransformer([
    ("num_transformer", make_pipeline(imputer, scaler), num_features),
    ("cat_transformer", cat_encoder, cat_features),
])

preprocessor
```

```python
preprocessor.fit(model_features)
prepared_features = pd.DataFrame(
    preprocessor.transform(model_features).toarray(), columns=preprocessor.get_feature_names_out()
)
prepared_features.head()
```

```python
n = 141
pca = PCA(n_components=n)
pca_data = pca.fit_transform(prepared_features)
pca_df = pd.DataFrame(pca_data, columns=[f"c{i}" for i in range(1, n+1)])
```

```python
neighbors_model.fit(pca_df)
neighbors = neighbors_model.kneighbors(pca_df, return_distance=False)
```

```python
neighbors_indexes = neighbors[:, 1:]
neighbors_indexes[:5]
```

```python
base_indexes = [[i] * 3 for i in pca_df.index]
base_indexes[:5]
```

```python
base_tracks = features.take(np.ravel(base_indexes)).reset_index(drop=True)
neighbor_tracks = features.take(np.ravel(neighbors_indexes)).reset_index(drop=True)
```

## Exps with distance metrics 

<span style="background-color: #990000">__Result__</span>:          - `manhattan`.

```python
mertics_types = [
    "euclidean", 
    "manhattan",
    "cosine",
    "minkowski",
]

df_metrics = pd.DataFrame()

for metric in mertics_types:
    neighbors_model = NearestNeighbors(n_neighbors=4, metric=metric)
    neighbors_model.fit(prepared_features)
    neighbors = neighbors_model.kneighbors(prepared_features, return_distance=False)
    neighbors_indexes = neighbors[:, 1:]
    
    base_indexes = [[i] * 3 for i in range(features.shape[0])]
    base_tracks = features.take(np.ravel(base_indexes))

    base_tracks = features.take(np.ravel(base_indexes)).reset_index(drop=True)
    neighbor_tracks = features.take(np.ravel(neighbors_indexes)).reset_index(drop=True)
    
    if df_metrics.empty:
        df_metrics = pd.DataFrame(
            {
                "metirc_method": metric,
                "genre_metric": [genre_metric(base_tracks, neighbor_tracks)],
                "genre_multiple_metric": [genre_multiple_metric(base_tracks, neighbor_tracks)],
                "album_metric": [album_metric(base_tracks, neighbor_tracks)],
                "artist_metric": [artist_metric(base_tracks, neighbor_tracks)],
                "year_metric": [year_metric(base_tracks, neighbor_tracks)], 
                "sound_metric": [sound_metric(base_tracks, neighbor_tracks)]
            }
        )
    else:
        df_tmp = pd.DataFrame(
            {
                "metirc_method": metric,
                "genre_metric": [genre_metric(base_tracks, neighbor_tracks)],
                "genre_multiple_metric": [genre_multiple_metric(base_tracks, neighbor_tracks)],
                "album_metric": [album_metric(base_tracks, neighbor_tracks)],
                "artist_metric": [artist_metric(base_tracks, neighbor_tracks)],
                "year_metric": [year_metric(base_tracks, neighbor_tracks)], 
                "sound_metric": [sound_metric(base_tracks, neighbor_tracks)]
            }
        )
        df_metrics = pd.concat([df_metrics, df_tmp])
```

```python
df_metrics.sort_values(by='genre_metric')
```

```python
df_mertics = pd.DataFrame()
for n in range(1, 147):
    pca = PCA(n_components=n)
    pca_data = pca.fit_transform(prepared_features)
    pca_df = pd.DataFrame(pca_data, columns=[f"c{i}" for i in range(1, n+1)])
    neighbors_model = NearestNeighbors(n_neighbors=4, metric="manhattan")
    neighbors_model.fit(pca_df)
    neighbors = neighbors_model.kneighbors(pca_df, return_distance=False)
    neighbors_indexes = neighbors[:, 1:]

    base_indexes = [[i] * 3 for i in range(features.shape[0])]
    base_tracks = features.take(np.ravel(base_indexes))

    base_tracks = features.take(np.ravel(base_indexes)).reset_index(drop=True)
    neighbor_tracks = features.take(np.ravel(neighbors_indexes)).reset_index(drop=True)

    df_tmp = pd.DataFrame({
        "n_components": [n],
        "genre_metric": [genre_metric(base_tracks, neighbor_tracks)],
        "genre_multiple_metric": [genre_multiple_metric(base_tracks, neighbor_tracks)],
        "album_metric": [album_metric(base_tracks, neighbor_tracks)],
        "artist_metric": [artist_metric(base_tracks, neighbor_tracks)],
        "year_metric": [year_metric(base_tracks, neighbor_tracks)], 
        "sound_metric": [sound_metric(base_tracks, neighbor_tracks)]
    })

    if df_mertics.empty:
        df_mertics = df_tmp
    else:
        df_mertics = pd.concat([df_mertics, df_tmp])
```
